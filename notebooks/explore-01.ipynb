```python
from pathlib import Path
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

datadir = Path('.').resolve().parents[0] / "data"
datadir
```

```python
df = pd.read_csv(datadir / "interim" / "wine_df_nice_cols.csv")
df.head()
```

```python
g = sns.FacetGrid(df, row="class",
                  height=1.7, aspect=4,)
g.map(sns.distplot, "alcohol", hist=False, rug=True);
```

```python
g = sns.PairGrid(df, hue="class")
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
```

```python
# zoom in
g = sns.PairGrid(df[df.columns[0:5]], hue="class", height=2.5)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
